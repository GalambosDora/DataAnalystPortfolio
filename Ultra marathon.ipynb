```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Filtering our data down to 50km or 50miles and year 2020

df[(df['Event distance/length'].isin(["50km","50mi"])) \
& (df['Year of event'] == 2020)]
```

```python
# getting country out of event name and filtering for USA

df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == "USA"]
```

```python
## combine all filters together: 50km/50mi, 2020, USA

df2 = df[(df['Event distance/length'].isin(["50km","50mi"])) \
& (df['Year of event'] == 2020) \
& (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == "USA")]
```

```python
# remove USA from event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
# cleaning age

df2['Athlete_age'] = 2020-df2['Athlete year of birth']
```

```python
# cleaning performance 

df2['Athlete performance']= df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
# droping unnecessary columns
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head()
```

```python
# cleaning NaN values

df2.isna().sum()
```

```python
df2[df2['Athlete_age'].isna() == 1]
```

```python
df2=df2.dropna()
```

```python
# reset index

df2.reset_index(drop=True)
```

```python
# changing datatypes

df2['Athlete_age'] = df2['Athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates': 'race_day', 
                            'Event name': 'race_name', 
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_number_of_finishers', 
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed':  'athlete_average_speed' , 
                            'Athlete ID': 'athlete_id', 
                            'Athlete_age': 'athlete_age'})
```

```python
# reordering columns

df3 = df2[['race_day','race_name','race_length','race_number_of_finishers', 'athlete_id', 'athlete_gender','athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
df3['race_number_of_finishers'].describe()
```

```python
# easier way of getting the name of race(s) with the most finishers

max_number = df3.query('race_number_of_finishers == 636') 
max_number['race_name'].unique().tolist()
```

```python
# more complex way of getting the name of race(s) with the most finishers

max_finisher = df3['race_number_of_finishers'].max()
result = df3.query('race_number_of_finishers == @max_finisher')
race_with_max_finishers = result['race_name'].unique().tolist()

print(race_with_max_finishers)
print("Number of max. finisher:", max_finisher)
```

```python
# easier way of getting the races with min number of finishers

min_number = df3.query('race_number_of_finishers == 2')

min_number['race_name'].unique().tolist()
```

```python
# more complex way of gettiong the name of race(s) with the smallest number of finishers

min_finisher = df3['race_number_of_finishers'].min()
result = df3.query('race_number_of_finishers == @min_finisher')
race_with_min_finishers = result['race_name'].unique().tolist()

print(race_with_min_finishers)
print("Number of min. finisher:", min_finisher)
```

```python
# number of males and females
women = df3.query("athlete_gender == 'F'")
no_of_women = women['athlete_id'].count()

print(no_of_women, "took part in the races")



men = df3.query("athlete_gender == 'M'")
no_of_men = men['athlete_id'].count()

print(no_of_men, "took part in the races")

print(no_of_men - no_of_women, "more men took part in the races than women")
```

```python
#fastest females

women = df3.query("athlete_gender == 'F'")
fastest_time = women['athlete_average_speed'].max()

print("The average speed of the fastest woman was", fastest_time)

fastest_athlete = women.query('athlete_average_speed == @fastest_time')
fastest_athlete_id = fastest_athlete['athlete_id'].values[0]

print("The id of the fastest woman was:", fastest_athlete_id)
```

```python
# searching for duplicates

df3[df3.duplicated(subset=['race_day','race_name', 'athlete_id', 'race_number_of_finishers', 'athlete_gender'], keep = False)]
```

```python
df4 = df3.drop_duplicates(subset = ['race_day','race_name', 'athlete_id', 'race_number_of_finishers', 'athlete_gender'], keep = 'last').reset_index(drop = True)
```

```python
# what age groups are the best in the 50mi race (age group with min 20 races)

result = df4.query('race_length == "50mi"').groupby('athlete_age')\
['athlete_average_speed'].agg(['mean', 'count'])\
.sort_values('mean', ascending = False).query('count>19')

result
```

```python
# which season are the racers the fastest
# creating race month column
df4['race_month'] = df4['race_day'].str.split('.').str.get(1).astype(int)
```

```python
# creating race season column
df4['race_season'] = df4['race_month'].apply(lambda x: "Winter" if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter' )
```

```python
df4.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
